```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.pandas.set_option("display.max_columns", None)
```

```python
data = pd.read_csv("Road Accident Data.csv")
```

```python
data.head(10)
```

```python
data.shape
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
data = data.drop("Carriageway_Hazards", axis = 1)
```

```python
data = data.dropna().reset_index()
```

```python
data.shape
```

```python
data.head()
```

```python
import datetime as dt
data["Accident Date"] = pd.to_datetime(data["Accident Date"])
data["Time"] = pd.to_datetime(data["Time"])
```

```python
data.head()
```

```python
data['Day_of_Week'] = data["Accident Date"].dt.day_name()
data['Day_no'] = data["Accident Date"].dt.dayofweek
data['day_of_year'] = data["Accident Date"].dt.dayofyear
```

```python
data.head()
```

```python
data['month'] = data["Accident Date"].dt.month
data['month_name'] = data["Accident Date"].dt.month_name()
```

```python
data["Hour_of_day"] = data["Time"].dt.hour
data.head()
```

```python
plt.figure(figsize = (10, 8))
sns.countplot(data = data, x = "Hour_of_day")
plt.show()
```

```python
#This plot shows that most accidents occured at 5pm
#Also between 10pm to 6am there are few number of accident case.
```

```python
plt.figure(figsize = (10, 8))
sns.countplot(data = data, x = "Day_of_Week")
plt.xticks(rotation = 45)
plt.show()
```

```python
##This plot shows that accidents mostly occured on saturday
```

```python
plt.figure(figsize = (10, 8))
sns.countplot(data = data, x = "month")
plt.xticks(rotation = 45)
plt.show()
```

```python
## This graph shows that most accidents occured in the 11th month
```

```python
plt.figure(figsize = (10, 8))
sns.countplot(data = data, x = "Light_Conditions", hue = "Accident_Severity")
plt.xticks(rotation = 45)
plt.show()
```

```python
#This plot shows that most accidents occured in daylight
#Also the plot shows that fatal, serious, and slight accidents occured mostly in daylight
#Fetal accidents didn't really occur
```

```python
plt.figure(figsize = (10, 8))
sns.countplot(data = data, hue = "Accident_Severity", x = "Weather_Conditions")
plt.xticks(rotation = 45)
plt.show()
```

```python
#This plot shows that most accidents occured under Fine no high winds conditon
#Also the plot shows that fatal, serious, and slight accidents occured mostly under Fine no high winds condition
#Slight Accident mostly occured
```

```python
plt.figure(figsize = (10, 8))
sns.countplot(data = data, x = "Vehicle_Type")
plt.xticks(rotation = 90)
plt.show()
```

```python
#This plot shows that most accident occurred by Car
```

```python
plt.figure(figsize = (10, 8))
sns.countplot(data = data, x = "Junction_Detail")
plt.xticks(rotation = 90)
plt.show()
```

```python
local_pop_data = data["Local_Authority_(District)"].value_counts()
local_pop_data = pd.DataFrame(local_pop_data)
local_pop_data_top_15 = local_pop_data.head(15)
local_pop_data_top_15
```

```python
plt.figure(figsize = (10, 8))
sns.barplot(data = local_pop_data_top_15, x = "Local_Authority_(District)", y = "count")
plt.xticks(rotation = 90)
plt.show()
```

```python
#This chart shows that Birmingham city recorded the most accident cases
```

```python
local_pop_data_buttom_15 = local_pop_data.tail(15)
local_pop_data_buttom_15
```

```python
plt.figure(figsize = (10, 8))
sns.barplot(data = local_pop_data_buttom_15, x = "Local_Authority_(District)", y = "count")
plt.xticks(rotation = 90)
plt.show()
```
