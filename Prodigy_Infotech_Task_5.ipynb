```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.pandas.set_option("display.max_columns", None)
```

```python
data = pd.read_csv("Road Accident Data.csv")
```

```python
data.head(10)
```

```python
data.shape
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
data = data.drop("Carriageway_Hazards", axis = 1)
```

```python
data = data.dropna().reset_index()
```

```python
data.shape
```

```python
data.head()
```

```python
data.dtypes
```

```python
import datetime as dt
data["Accident Date"] = pd.to_datetime(data["Accident Date"])
data["Time"] = data["Time"].astype("object")
```

```python
data["Hour"] = data["Time"].str.split(":").str[0].astype("int")
data["Minute"] = data["Time"].str.split(":").str[1].astype("int")
```

```python
data.head()
```

```python
data['Day_of_Week'] = data["Accident Date"].dt.day_name()
data['Day_no'] = data["Accident Date"].dt.dayofweek
data['day_of_year'] = data["Accident Date"].dt.dayofyear
```

```python
data.head()
```

```python
data['month'] = data["Accident Date"].dt.month
data['month_name'] = data["Accident Date"].dt.month_name()
```

```python
data = data.drop(["Time", "Accident Date"], axis = 1)
data.dtypes
```

```python
plt.figure(figsize = (10, 8))
sns.countplot(data = data, x = "Hour")
plt.show()
```

## This plot shows that most accidents occured at 5pm
## Also between 10pm to 6am there are few number of accident case.

```python
plt.figure(figsize = (10, 8))
sns.countplot(data = data, x = "Day_of_Week")
plt.xticks(rotation = 45)
plt.show()
```

## This plot shows that accidents mostly occured on Saturday, and least on Mondays

```python
plt.figure(figsize = (10, 8))
sns.countplot(data = data, x = "month")
plt.xticks(rotation = 45)
plt.show()
```
